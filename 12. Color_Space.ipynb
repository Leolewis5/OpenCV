直接引述維基百科的文章  
大多數電視機、顯示器、投影儀通過將不同強度的紅、綠、藍色光混合來生成不同的顏色，這就是RGB 三原色的加色法。通過這種方法可以在RGB色彩空間生成大量不同的顏色，然而，這三種顏色分量的取值與所生成的顏色之間的聯繫並不直觀。

藝術家有時偏好使用HSL或HSV而不選擇三原色光模式（即RGB模型）或 印刷四分色模式（即CMYK模型），因為它類似於人類感覺顏色的方式，具有較強的感知度。RGB和CMYK分別是加法原色和減法原色模型，以原色組合的方式定義顏色，而HSV以人類更熟悉的方式封裝了關於顏色的信息：「這是什麼顏色？深淺如何？明暗如何？」。

```python
import os
import cv2 as cv

root = os.getcwd()
img_path = os.path.join(root,'Image/lena.jpg')
img = cv.imread(img_path,1)
```

先定義一個把三個通道的圖像畫出來的簡單函示

```python
import matplotlib.pyplot as plt

def plot3channel(img,string):
    string = string.split('.')
    plt.figure(figsize=(15,5))
    plt.subplot(131),plt.imshow(img[:,:,0],'gray'),plt.title('channel: '+ string[0]),plt.colorbar()
    plt.subplot(132),plt.imshow(img[:,:,1],'gray'),plt.title('channel: '+ string[1]),plt.colorbar()
    plt.subplot(133),plt.imshow(img[:,:,2],'gray'),plt.title('channel: '+ string[2]),plt.colorbar()
```

## RGB space

RGB空間中三個通道的灰階顯示  

```python
plt.imshow(img[:,:,::-1])
plot3channel(img,'B.G.R')
```

## HSV space

* 色相（H）是色彩的基本屬性，就是平常所說的顏色名稱，如紅色、黃色等。
* 飽和度（S）是指色彩的純度，越高色彩越純，低則逐漸變灰，取0-100%的數值。
* 明度（V），亮度（L），取0-100%。

```python
img2= cv.cvtColor(img,cv.COLOR_BGR2HSV)

plt.imshow(img2[:,:,::-1])
plot3channel(img2,'H.S.V')
```

## HLS

![](https://upload.wikimedia.org/wikipedia/commons/thumb/b/b6/HSL_HSV_cylinder_color_solid_comparison.png/400px-HSL_HSV_cylinder_color_solid_comparison.png)

HSV中的S控制純色中混入白色的量，值越大，白色越少，顏色越純;  
HSV中的V控制純色中混入黑色的量，值越大，黑色越少，明度越高  
HSL中的 S和黑白沒有關係，飽和度不控制顏色中混入黑白的多寡;  
HSL中的L控制純色中的混入的黑白兩種顏色。

```python
img3= cv.cvtColor(img,cv.COLOR_BGR2HLS)

plt.imshow(img3[:,:,::-1])
plot3channel(img3,'H.L.S')
```
