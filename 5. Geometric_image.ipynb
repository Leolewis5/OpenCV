# Geometric image

幾何圖像的繪製  
幾何圖象就是：線、矩形、圓形、三角形等等  
要繪製幾何圖象需要的就是座標點，給訂做標點  
就可以在該座標點上標上顏色，進而形成圖像  
opencv給予了一些函示可以用來畫出基本的幾何圖象

## Straight line
直線  
函式為：cv.line（img,Point pt1,Point pt2,color,thickness=1,line_type=8, shift=0)

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((256,256),np.uint8)# create img
cv.line(img,(0,0),(256,256),255,3,3,0)


plt.imshow(img,'gray')
```

## Rectangle
矩形
函式為：cv.rectangle（img,Point pt1,Point pt2,color,thickness=1,line_type=8 shift=0)  
跟線段一樣，給定兩點  

```python
img = np.zeros((256,256),np.uint8)# create img
cv.rectangle(img,(50,50),(200,200),255,3,3,0)

plt.imshow(img,'gray')
```

## Circle
圓形
函式為：cv.circle （img,center point,radius,color,thickness=1,line_type=8 shift=0)  
非常直觀，給定圓心與半徑即可  
在thichness給予-1可以畫出實心圖

```python
img = np.zeros((256,256),np.uint8)# create img
cv.circle(img,(128,128),50,255,3,3,0)


img2 = np.zeros((256,256),np.uint8)# create img
cv.circle(img2,(128,128),50,255,-1,3,0)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1),plt.imshow(img,'gray')
plt.subplot(1,2,2),plt.imshow(img2,'gray')     
```

# Oval

橢圓，較為複雜  
函式為ellipse，需要代入以下參數  
ellipse(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType[, shift]]])  
* img: 畫布
* center: 圓心座標
* axes: 軸長
* angle: 旋轉角度
* startAngle: 圓弧起始角度
* endAngle: 圓弧終結角度
* color: 顏色
* thickness: 厚度，給予-1可以讓其變為實心

```python
img = np.zeros((256,256),np.uint8)# create img
cv.ellipse(img,(128,128),(80,55),30,0,270,255,-1,8)


img2 = np.zeros((256,256),np.uint8)# create img
cv.ellipse(img2,(128,128),(80,55),-30,0,220,255,3,8)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1),plt.imshow(img,'gray')
plt.subplot(1,2,2),plt.imshow(img2,'gray')     
```
